### Listings

```python
query_listings = """select * from listings"""
```

```python
df_data_listings = pd.read_sql(query_listings, conn)
```

```python
df_data_listings.shape
```

```python
df_data_listings.head(3)
```

```python
# Delete Dollar Sign ($) and Comma (,)

# without missing value
df_data_listings['price'] = df_data_listings['price'].apply(lambda x: x.lstrip('$')).str.replace(',', '')
df_data_listings['extra_people'] = df_data_listings['extra_people'].apply(lambda x: x.lstrip('$')).str.replace(',', '')

# with missing value
df_data_listings['weekly_price'] = df_data_listings['weekly_price'].map(lambda x: x and x.lstrip('$')).str.replace(',', '')
df_data_listings['monthly_price'] = df_data_listings['monthly_price'].map(lambda x: x and x.lstrip('$')).str.replace(',', '')
df_data_listings['security_deposit'] = df_data_listings['security_deposit'].map(lambda x: x and x.lstrip('$')).str.replace(',', '')
df_data_listings['cleaning_fee'] = df_data_listings['cleaning_fee'].map(lambda x: x and x.lstrip('$')).str.replace(',', '')
```

```python
df_data_listings.head(3)
```

```python
# Change Data Type to Float
df_data_listings['price'] = df_data_listings['price'].astype('float')
df_data_listings['weekly_price'] = df_data_listings['weekly_price'].astype('float')
df_data_listings['monthly_price'] = df_data_listings['monthly_price'].astype('float')
df_data_listings['security_deposit'] = df_data_listings['security_deposit'].astype('float')
df_data_listings['cleaning_fee'] = df_data_listings['cleaning_fee'].astype('float')
df_data_listings['extra_people'] = df_data_listings['extra_people'].astype('float')
```

```python
# Change USD to SGP
df_data_listings['price'] = df_data_listings['price'] * 1.37092
df_data_listings['weekly_price'] = df_data_listings['weekly_price'] * 1.37092
df_data_listings['monthly_price'] = df_data_listings['monthly_price'].astype('float') * 1.37092
df_data_listings['security_deposit'] = df_data_listings['security_deposit'].astype('float') * 1.37092
df_data_listings['cleaning_fee'] = df_data_listings['cleaning_fee'].astype('float') * 1.37092
df_data_listings['extra_people'] = df_data_listings['extra_people'].astype('float') * 1.37092
```

```python
df_data_listings.head(3)
```

```python
# Change Last Review to Date Time
df_data_listings['last_review'] = pd.to_datetime(df_data_listings['last_review'], errors='coerce')
```

```python
df_data_listings['last_review'].head()
```

```python
# Filter Last Review
filter = df_data_listings['last_review'] < '2016-08-01'
df_data_listings = df_data_listings[filter]
```

```python
df_data_listings.shape
```

```python
df_data_listings.head(2)
```

```python
df_data_listings.drop(['listing_url','thumbnail_url','medium_url','picture_url','xl_picture_url','host_url',
                       'host_thumbnail_url','host_picture_url'], axis=1, inplace=True)
```

```python
df_data_listings.shape
```

```python
df_data_listings.head(2)
```

```python
# Push Data Listings into Data Warehouse (DWH)
df_data_listings.to_sql(con=conn_dwh, name='LISTING_via_731', if_exists='replace', index=False)
```
